```python
# fit and evaluate a model
def evaluate_model(trainX, trainy, testX, testy):
    verbose, epochs, batch_size = 0, 10, 32
    n_timesteps, n_features, n_outputs = trainX.shape[1], trainX.shape[2], trainy.shape[1]
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_timesteps,n_features)))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(Dropout(0.5))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    # fit network
    model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, verbose=verbose)
    # evaluate model
    _, accuracy = model.evaluate(testX, testy, batch_size=batch_size, verbose=0)
    return accuracy
```

```python
def compare_all_feature_importance(df, x_columns, y_columns, title=None):
    
    def get_all_importances(df, x_columns, y_columns):
        importances = []
        rf = Data.models[4][1]
        for y_col in y_columns:
            try:
                X,y = get_X_y(df, x_columns, y_col)
                rf.fit(X,y)
                importances.append(rf.feature_importances_)
                # print(y_col)
            except ValueError:
                pass
        return importances
    
    def reshape_importances(importances, num_features = 4):
        reshaped_importances = [[] for _ in range(num_features)]
        for fi in importances:
            for i in range(num_features):
                value = fi[i]
                reshaped_importances[i].append(value)
        return reshaped_importances
    
    def plot_feature_importance(reshaped_importances, title=None):
        pd.DataFrame({
            'Feature':['Temp', 'WindX', 'WindY', 'Pres'],
            'Importance':[np.array(x).mean() for x in reshaped_importances],
            'StdDev':[np.array(x).std() for x in reshaped_importances],

        }).plot.bar(x='Feature', title=title)
        pyplot.show()
        
    importances = get_all_importances(df, x_columns, y_columns)
    reshaped_importances = reshape_importances(importances, len(x_columns))
    plot_feature_importance(reshaped_importances, title)
    
compare_all_feature_importance(Data.delta_time,Data.delta_time.columns[:4], Data.delta_time.columns[5:], title="Feature Importance: Yearly Dataset")
compare_all_feature_importance(Data.delta2_time,Data.delta2_time.columns[:4], Data.delta2_time.columns[5:], title="Feature Importance: Monthly Dataset")
```

```python
def compare_all_stations(df, x_columns, y_columns, title=None):
    def plot(model_scores, model_stds, title=None):
        errors = []
        names = []
        stds = []

        for (name,_),error,std in zip(Data.models, [abs(sum(x))/len(x) for x in model_scores], [sum(x)/len(x) for x in model_stds]):
            names.append(name)
            errors.append(error)
            stds.append(std)

        pd.DataFrame({
            'Model':names,
            'Error':errors,
            'StDev':stds,

        }).sort_values(by=['Error','StDev']).plot.bar(x='Model', title=title)
        pyplot.show()
        
    model_scores = [[] for i in Data.models]
    model_stds = [[] for i in Data.models]

    for X,y in (get_X_y(df, x_columns, y_col) for y_col in y_columns):
        try:
            for j, (name, model) in enumerate(Data.models):
                dic = cross_validate(model, X, y, error_score='raise',cv=KFold(n_splits=5).split(X), scoring='neg_mean_squared_error',return_estimator=True)
                model_scores[j].append(dic['test_score'].mean())
                model_stds[j].append(dic['test_score'].std())

        except ValueError:
            pass
        
    plot(model_scores, model_stds, title)
    
compare_all_stations(Data.delta_time, Data.delta_time.columns[:4], Data.delta_time.columns[5:], 'Yearly All Stations')
compare_all_stations(Data.delta2_time, Data.delta2_time.columns[:4], Data.delta2_time.columns[5:], 'Monthly All Stations')
```
