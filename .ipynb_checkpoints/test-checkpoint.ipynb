```python
string = '30AB, 39A, 40AB, 46AB, 48-50 AB, 55 AB, 58 AB, 66-67 AB, 71-74 AB'
```

```python
string.split(', ')
```

```python
def generate_from_string(string):
    def hasB(num, station):
        if 'A' in station:
            yield 5+2*(num-1)
        if 'B' in station:
            yield 6+2*(num-1)
    
    
    for station in string.split(', '):
        if '-' in station:
            for i in range(int(station[:2]), int(station[3:5])+1):
                yield from hasB(i, station)
        else:
            yield from hasB(int(station[:2]), station)

print(string)
for i in generate_from_string(string):

    print(Data.delta_time.columns[i])
    print()
    
```
